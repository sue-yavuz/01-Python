1. Two Sum

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        nums2 = nums.copy()
        for i in nums:
            nums2.remove(i)
            for j in nums2:
                if i+j == target:
                    x = nums.index(i)
                    y = nums.index(j, x+1)
                    return [x, y]
```

2. Add Two Numbers

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        head  = None
        temp  = None
        carry  = 0

        while l1 is not None or l2 is not None:

            sum = carry

            if l1 is not None:
                sum += l1.val
                l1 = l1.next

            if l2 is not None:
                sum += l2.val
                l2 = l2.next

            carry = sum // 10
            node = ListNode(sum % 10)

            if temp is None:
                temp = head = node
            else:
                temp.next = node
                temp      = temp.next

        if carry > 0:
            temp.next = ListNode(carry)

        return head
```

```python
# 1480. Running Sum of 1d Array

class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        output = []
        
        for i in range(1,len(nums)+1):
            result = sum(nums[:i])
            output.append(result)

        return output
```

```python
# 724. Find Pivot Index
# 1.way 
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        for i in range(len(nums)):
            leftside = sum(nums[:i])
            rigthside = sum(nums[i+1:])

            if leftside == rigthside:
                return i 
        return -1

#2.way 
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        s, presum = sum(nums), 0
        for i, v in enumerate(nums):
            if (presum << 1) == s - v:
                return i
            presum += v
        return -1
```
